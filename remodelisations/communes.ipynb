```python
import pandas as pd
import numpy as np
import os
```

```python
folder = './csv'
```

```python
csv_file = os.path.join(folder, 'commune2021.csv')
csv_file

folder = './csv/filtered'
```

```python
data = pd.read_csv(csv_file, sep=',')
data
```

```python
df = data[data['TYPECOM'] == 'COM']
df = df[["COM", "REG", "DEP", "NCC", "LIBELLE"]]
df['REG'] = df['REG'].apply(lambda x: int(x))

df
```

```python
new_csv = os.path.join(folder, 'communes.csv')
df.to_csv(new_csv, ',', index=False, header=False)
```
